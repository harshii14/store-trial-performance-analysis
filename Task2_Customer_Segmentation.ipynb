```python

df = pd.read_csv('QVI_data.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
df['MONTH'] = df['DATE'].dt.to_period('M').astype(str)
monthly_metrics = df.groupby(['STORE_NBR', 'MONTH']).agg(
    total_sales=('TOT_SALES', 'sum'),
    n_customers=('LYLTY_CARD_NBR', pd.Series.nunique),
    n_txn=('TXN_ID', 'nunique')
).reset_index()

monthly_metrics['avg_txn_per_customer'] = monthly_metrics['n_txn'] / monthly_metrics['n_customers']
def find_control_store(trial_store, metric='total_sales'):
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store]
    stores = monthly_metrics['STORE_NBR'].unique()
    stores = stores[stores != trial_store]

    scores = []

    for store in stores:
        store_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store]
        merged = pd.merge(trial_data, store_data, on='MONTH', suffixes=('_trial', '_control'))

        # Correlation calculation only if variation exists
        if len(merged) > 0 and merged[f'{metric}_trial'].std() > 0 and merged[f'{metric}_control'].std() > 0:
            corr = merged[f'{metric}_trial'].corr(merged[f'{metric}_control'])
        else:
            corr = -1

        # Calculate magnitude distance
        dist = np.mean(np.abs(merged[f'{metric}_trial'] - merged[f'{metric}_control']))
        max_dist = merged[f'{metric}_trial'].max() - merged[f'{metric}_trial'].min()
        similarity_score = (1 - dist / max_dist) if max_dist != 0 else 0

        overall_score = (corr + similarity_score) / 2
        scores.append((store, overall_score))

    best_match = sorted(scores, key=lambda x: x[1], reverse=True)[0]
    return best_match[0]
def find_control_store(trial_store, metric='total_sales'):
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store]
    stores = monthly_metrics['STORE_NBR'].unique()
    stores = stores[stores != trial_store]

    scores = []

    for store in stores:
        store_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store]
        merged = pd.merge(trial_data, store_data, on='MONTH', suffixes=('_trial', '_control'))

        # Correlation calculation only if variation exists
        if len(merged) > 0 and merged[f'{metric}_trial'].std() > 0 and merged[f'{metric}_control'].std() > 0:
            corr = merged[f'{metric}_trial'].corr(merged[f'{metric}_control'])
        else:
            corr = -1

        # Calculate magnitude distance
        dist = np.mean(np.abs(merged[f'{metric}_trial'] - merged[f'{metric}_control']))
        max_dist = merged[f'{metric}_trial'].max() - merged[f'{metric}_trial'].min()
        similarity_score = (1 - dist / max_dist) if max_dist != 0 else 0

        overall_score = (corr + similarity_score) / 2
        scores.append((store, overall_score))

    best_match = sorted(scores, key=lambda x: x[1], reverse=True)[0]
    return best_match[0]
def find_control_store(trial_store, metric='total_sales'):
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store]
    stores = monthly_metrics['STORE_NBR'].unique()
    stores = stores[stores != trial_store]

    scores = []

    for store in stores:
        store_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store]
        merged = pd.merge(trial_data, store_data, on='MONTH', suffixes=('_trial', '_control'))

        # Correlation calculation only if variation exists
        if len(merged) > 0 and merged[f'{metric}_trial'].std() > 0 and merged[f'{metric}_control'].std() > 0:
            corr = merged[f'{metric}_trial'].corr(merged[f'{metric}_control'])
        else:
            corr = -1

        # Calculate magnitude distance
        dist = np.mean(np.abs(merged[f'{metric}_trial'] - merged[f'{metric}_control']))
        max_dist = merged[f'{metric}_trial'].max() - merged[f'{metric}_trial'].min()
        similarity_score = (1 - dist / max_dist) if max_dist != 0 else 0

        overall_score = (corr + similarity_score) / 2
        scores.append((store, overall_score))

    best_match = sorted(scores, key=lambda x: x[1], reverse=True)[0]
    return best_match[0]
def find_control_store(trial_store, metric='total_sales'):
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store]
    stores = monthly_metrics['STORE_NBR'].unique()
    stores = stores[stores != trial_store]

    scores = []

    for store in stores:
        store_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store]
        merged = pd.merge(trial_data, store_data, on='MONTH', suffixes=('_trial', '_control'))

        # Correlation calculation only if variation exists
        if len(merged) > 0 and merged[f'{metric}_trial'].std() > 0 and merged[f'{metric}_control'].std() > 0:
            corr = merged[f'{metric}_trial'].corr(merged[f'{metric}_control'])
        else:
            corr = -1

        # Calculate magnitude distance
        dist = np.mean(np.abs(merged[f'{metric}_trial'] - merged[f'{metric}_control']))
        max_dist = merged[f'{metric}_trial'].max() - merged[f'{metric}_trial'].min()
        similarity_score = (1 - dist / max_dist) if max_dist != 0 else 0

        overall_score = (corr + similarity_score) / 2
        scores.append((store, overall_score))

    best_match = sorted(scores, key=lambda x: x[1], reverse=True)[0]
    return best_match[0]
def find_control_store(trial_store, metric='total_sales'):
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store]
    stores = monthly_metrics['STORE_NBR'].unique()
    stores = stores[stores != trial_store]

    scores = []

    for store in stores:
        store_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store]
        merged = pd.merge(trial_data, store_data, on='MONTH', suffixes=('_trial', '_control'))

        # Correlation calculation only if variation exists
        if len(merged) > 0 and merged[f'{metric}_trial'].std() > 0 and merged[f'{metric}_control'].std() > 0:
            corr = merged[f'{metric}_trial'].corr(merged[f'{metric}_control'])
        else:
            corr = -1

        # Calculate magnitude distance
        dist = np.mean(np.abs(merged[f'{metric}_trial'] - merged[f'{metric}_control']))
        max_dist = merged[f'{metric}_trial'].max() - merged[f'{metric}_trial'].min()
        similarity_score = (1 - dist / max_dist) if max_dist != 0 else 0

        overall_score = (corr + similarity_score) / 2
        scores.append((store, overall_score))

    best_match = sorted(scores, key=lambda x: x[1], reverse=True)[0]
    return best_match[0]
trial_stores = [77, 86, 88]
control_stores = {}

for trial in trial_stores:
    control = find_control_store(trial)
    control_stores[trial] = control

control_stores
def plot_comparison(trial, control):
    subset = monthly_metrics[monthly_metrics['STORE_NBR'].isin([trial, control])]
    pivot = subset.pivot(index='MONTH', columns='STORE_NBR', values='total_sales')

    pivot.plot(title=f'Store {trial} vs Store {control} - Total Sales', figsize=(12, 5))
    plt.ylabel("Total Sales")
    plt.show()

    print(f"Avg total sales difference: {pivot[trial].mean() - pivot[control].mean():.2f}")
plot_comparison(77, control_stores[77])
plot_comparison(86, control_stores[86])
plot_comparison(88, control_stores[88])
monthly_metrics.to_csv('monthly_metrics.csv', index=False)

```
